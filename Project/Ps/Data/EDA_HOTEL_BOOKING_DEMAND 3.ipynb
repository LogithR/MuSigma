```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Replace 'your_dataset.csv' with the actual file name
data = pd.read_csv(('Hotel_Bookings.csv'))

# Convert 'arrival_date_year', 'arrival_date_month' to a datetime format
data['arrival_date'] = pd.to_datetime(data['arrival_date_year'].astype(str) + '-' +
                                       data['arrival_date_month'] + '-' +
                                       data['arrival_date_day_of_month'].astype(str))

# Extract month and year from the 'arrival_date'
data['month'] = data['arrival_date'].dt.month
data['year'] = data['arrival_date'].dt.year

# Aggregate data by month and year
monthly_bookings = data.groupby(['year', 'month']).size().reset_index(name='total_bookings')
monthly_cancellations = data[data['is_canceled'] == 1].groupby(['year', 'month']).size().reset_index(name='total_cancellations')

# Merge the two dataframes to have both bookings and cancellations in one dataframe
summary_df = pd.merge(monthly_bookings, monthly_cancellations, on=['year', 'month'], how='left')
summary_df['total_cancellations'] = summary_df['total_cancellations'].fillna(0)

# Pivot the data for better visualization
pivot_df = summary_df.pivot('month', 'year', 'total_bookings')
pivot_df_cancellations = summary_df.pivot('month', 'year', 'total_cancellations')

# Plot the number of bookings by month and year
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Total Bookings by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Plot the number of cancellations by month and year
plt.subplot(1, 2, 2)
sns.heatmap(pivot_df_cancellations, annot=True, cmap='YlOrRd', fmt='g')
plt.title('Total Cancellations by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Replace 'your_dataset.csv' with the actual file name
data = pd.read_csv('Hotel_Bookings.csv')

# Convert 'arrival_date_year', 'arrival_date_month' to a datetime format
data['arrival_date'] = pd.to_datetime(data['arrival_date_year'].astype(str) + '-' +
                                       data['arrival_date_month'] + '-' +
                                       data['arrival_date_day_of_month'].astype(str))

# Extract month and year from the 'arrival_date'
data['month'] = data['arrival_date'].dt.month
data['year'] = data['arrival_date'].dt.year

# Group data by year, month, and hotel type, and calculate average ADR
adr_summary = data.groupby(['year', 'month', 'hotel'])['adr'].mean().reset_index()

# Calculate average ADR for each month across all years for each hotel type
average_adr_by_month = adr_summary.groupby(['month', 'hotel'])['adr'].mean().reset_index()

# Find the month with the lowest average ADR for each hotel type
best_months = average_adr_by_month.loc[average_adr_by_month.groupby('hotel')['adr'].idxmin()]

# Create a pivot table for easier plotting
pivot_adr = adr_summary.pivot_table(index=['month', 'year'], columns='hotel', values='adr')

# Plot the average ADR by month and year for each type of hotel
plt.figure(figsize=(14, 7))
sns.lineplot(data=adr_summary, x='month', y='adr', hue='hotel', style='year', markers=True, dashes=False)

# Add markers for the best months
for hotel in best_months['hotel'].unique():
    hotel_data = best_months[best_months['hotel'] == hotel]
    for _, row in hotel_data.iterrows():
        plt.scatter(row['month'], row['adr'], label=f'{hotel} Best Time ({row["month"]})', edgecolor='black', s=100, zorder=5)

plt.title('Average Daily Rate (ADR) by Month, Year, and Hotel Type with Best Booking Times')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend(title='Hotel Type / Best Time', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Replace 'your_dataset.csv' with the actual file name
data = pd.read_csv('Hotel_Bookings.csv')

# Compute the total length of stay
data['total_stay_nights'] = data['stays_in_weekend_nights'] + data['stays_in_week_nights']

# Inspect data for missing or zero ADR values
print(data[['total_stay_nights', 'adr']].describe())
print(data[data['adr'] <= 0].groupby('total_stay_nights')['adr'].count())
print(data[data['total_stay_nights'] == 38]['adr'].describe())

# Remove rows where ADR is zero or less
data_cleaned = data[data['adr'] > 0]

# Further clean if needed: remove lengths of stay with very few records
stay_counts = data_cleaned['total_stay_nights'].value_counts()
valid_stay_lengths = stay_counts[stay_counts > 5].index
data_cleaned = data_cleaned[data_cleaned['total_stay_nights'].isin(valid_stay_lengths)]

# Group by total length of stay and calculate average ADR
adr_by_stay_length = data_cleaned.groupby('total_stay_nights')['adr'].mean().reset_index()

# Find the optimal length of stay with the lowest ADR
optimal_length_of_stay = adr_by_stay_length.loc[adr_by_stay_length['adr'].idxmin()]

# Plot ADR vs. Length of Stay
plt.figure(figsize=(10, 6))
sns.lineplot(data=adr_by_stay_length, x='total_stay_nights', y='adr', marker='o')

# Highlight the optimal length of stay
plt.scatter(optimal_length_of_stay['total_stay_nights'], optimal_length_of_stay['adr'],
            color='red', s=100, edgecolor='black', zorder=5, label=f'Optimal Stay: {optimal_length_of_stay["total_stay_nights"]} nights')

plt.title('Average Daily Rate (ADR) vs. Length of Stay')
plt.xlabel('Total Length of Stay (Nights)')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Hotel_Bookings.csv')

# Feature Engineering
# Create a column for total stay nights
data['total_stay_nights'] = data['stays_in_weekend_nights'] + data['stays_in_week_nights']

# Create a column for total number of people
data['total_people'] = data['adults'] + data['children'] + data['babies']

# Plot histograms with improved scaling

plt.figure(figsize=(18, 6))

# Total Stay Nights vs. Number of Special Requests
plt.subplot(1, 3, 1)
sns.histplot(data=data, x='total_stay_nights', hue='total_of_special_requests', multiple='stack', palette='viridis', bins=30)
plt.title('Total Stay Nights vs. Special Requests')
plt.xlabel('Total Stay Nights')
plt.ylabel('Count')
plt.yscale('log')  # Apply log scale for better visibility if counts are large

# Total Number of People vs. Number of Special Requests
plt.subplot(1, 3, 2)
sns.histplot(data=data, x='total_people', hue='total_of_special_requests', multiple='stack', palette='viridis', bins=30)
plt.title('Total Number of People vs. Special Requests')
plt.xlabel('Total Number of People')
plt.ylabel('Count')
plt.yscale('log')  # Apply log scale for better visibility if counts are large

# Meal Type vs. Number of Special Requests
plt.subplot(1, 3, 3)
sns.histplot(data=data, x='meal', hue='total_of_special_requests', multiple='stack', palette='viridis', discrete=True)
plt.title('Meal Type vs. Special Requests')
plt.xlabel('Meal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```
