## Replacing the Null Values with Mode

```python
most_frequent_agentid = a['agent'].mode()[0]
most_frequent_companyid = a['company'].mode()[0]
most_frequent_country = a['country'].mode()[0]
most_frequent_children = a['children'].mode()[0]

a['agent'].fillna(most_frequent_agentid, inplace=True)
a['children'].fillna(most_frequent_children, inplace=True)
a['company'].fillna(most_frequent_companyid, inplace=True)
a['country'].fillna(most_frequent_country, inplace=True)
```

## Converting the Datatype

```python
a['children'] = pd.to_numeric(a['children'], errors='coerce')
a['adr'] = pd.to_numeric(a['adr'], errors='coerce')
a['agent'] = pd.to_numeric(a['agent'], errors='coerce')
a['company'] = pd.to_numeric(a['company'], errors='coerce')

a['children'] = a['children'].astype('int64')
a['adr'] = a['adr'].astype('int64')
a['agent'] = a['agent'].astype('int64')
a['company'] = a['company'].astype('int64')
```

## Hotel Demand by Month

```python
import pandas as pd
import matplotlib.pyplot as plt

a['arrival_date_year'] = a['arrival_date_year'].astype(int)
a['arrival_date_month'] = a['arrival_date_month'].astype(str)
a['arrival_date_day_of_month'] = a['arrival_date_day_of_month'].astype(int)

a['arrival_date'] = pd.to_datetime(
    a[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].astype(str).agg('-'.join, axis=1),
    format='%Y-%B-%d' 
)

demand_by_month = a.groupby(a['arrival_date'].dt.to_period('M')).size()

peak_month = demand_by_month.idxmax()
peak_demand = demand_by_month.max()

plt.figure(figsize=(12, 6))
colors = ['red' if month == peak_month else 'blue' for month in demand_by_month.index]
demand_by_month.plot(kind='bar', color=colors)

plt.title('Hotel Demand by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')

plt.legend(handles=[plt.Line2D([0], [0], color='red', lw=4, label='Peak Month'),
                    plt.Line2D([0], [0], color='blue', lw=4, label='Other Months')])

plt.show()
```

## Average Daily Rate (ADR) by Month, Year, and Hotel Type with Best Booking Times

```python
a[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']] = a[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].apply(pd.to_numeric, errors='coerce')

a['arrival_date_str'] = a['arrival_date_year'].astype(str) + '-' + a['arrival_date_month'].astype(str).str.zfill(2) + '-' + a['arrival_date_day_of_month'].astype(str).str.zfill(2)
a['arrival_date'] = pd.to_datetime(a['arrival_date_str'], errors='coerce')

a['year_month'] = a['arrival_date'].dt.to_period('M').astype(str)

a = a.dropna(subset=['adr', 'is_repeated_guest', 'total_of_special_requests', 'market_segment', 'reservation_status'])

heatmap_data = a.pivot_table(index='year_month', columns='is_repeated_guest', values='adr', aggfunc='mean')

plt.figure(figsize=(14, 7))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Heatmap of Average Daily Rate (ADR) for Guest Type')
plt.xlabel('Repeated Guest (1 = Yes, 0 = No)')
plt.ylabel('Month-Year')
plt.xticks(ticks=[0.5, 1.5], labels=['New Guest', 'Repeat Guest'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
def plot_average_special_requests_by_room_type_sorted(a):
    # Calculate the average number of special requests for each room type
    room_type_requests_avg = a.groupby('reserved_room_type')['total_of_special_requests'].mean().reset_index()
   
    # Sort room types by average special requests in ascending order
    room_type_requests_avg_sorted = room_type_requests_avg.sort_values(by='total_of_special_requests', ascending=True)
   
    # Visualization of average special requests by room type
    plt.figure(figsize=(12, 8))
    sns.barplot(x='reserved_room_type', y='total_of_special_requests', data=room_type_requests_avg_sorted, color='skyblue')
    plt.title('Average Special Requests by Room Type (Ascending Order)')
    plt.xlabel('Room Type')
    plt.ylabel('Average Special Requests')
    plt.xticks(rotation=45)
    plt.show()
 
plot_average_special_requests_by_room_type_sorted(a)
```

```python
import warnings as wr
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a=pd.read_csv('Hotel_Bookings.csv')
wr.filterwarnings('ignore')

a['arrival_date'] = pd.to_datetime(a['arrival_date_year'].astype(str) + '-' +
                                       a['arrival_date_month'] + '-' +
                                       a['arrival_date_day_of_month'].astype(str))

a['month'] = a['arrival_date'].dt.month
a['year'] = a['arrival_date'].dt.year

monthly_bookings = a.groupby(['year', 'month']).size().reset_index(name='total_bookings')
monthly_cancellations = a[a['is_canceled'] == 1].groupby(['year', 'month']).size().reset_index(name='total_cancellations')

summary_df = pd.merge(monthly_bookings, monthly_cancellations, on=['year', 'month'], how='left')
summary_df['total_cancellations'] = summary_df['total_cancellations'].fillna(0)

pivot_df = summary_df.pivot('month', 'year', 'total_bookings')
pivot_df_cancellations = summary_df.pivot('month', 'year', 'total_cancellations')

# Plot the number of bookings by month and year
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.heatmap(pivot_df, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Total Bookings by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Plot the number of cancellations by month and year
plt.subplot(1, 2, 2)
sns.heatmap(pivot_df_cancellations, annot=True, cmap='YlOrRd', fmt='g')
plt.title('Total Cancellations by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

plt.tight_layout()
plt.show()
```

```python
a['arrival_date'] = pd.to_datetime(a['arrival_date_year'].astype(str) + '-' +
                                       a['arrival_date_month'] + '-' +
                                       a['arrival_date_day_of_month'].astype(str))

a['month'] = a['arrival_date'].dt.month
a['year'] = a['arrival_date'].dt.year

# Group data by year, month, and hotel type, and calculate average ADR
adr_summary = a.groupby(['year', 'month', 'hotel'])['adr'].mean().reset_index()

# Calculate average ADR for each month across all years for each hotel type
average_adr_by_month = adr_summary.groupby(['month', 'hotel'])['adr'].mean().reset_index()

# Find the month with the lowest average ADR for each hotel type
best_months = average_adr_by_month.loc[average_adr_by_month.groupby('hotel')['adr'].idxmin()]

# Create a pivot table for easier plotting
pivot_adr = adr_summary.pivot_table(index=['month', 'year'], columns='hotel', values='adr')

# Plot the average ADR by month and year for each type of hotel
plt.figure(figsize=(14, 7))
sns.lineplot(data=adr_summary, x='month', y='adr', hue='hotel', style='year', markers=True, dashes=False)

# Add markers for the best months
for hotel in best_months['hotel'].unique():
    hotel_data = best_months[best_months['hotel'] == hotel]
    for _, row in hotel_data.iterrows():
        plt.scatter(row['month'], row['adr'], label=f'{hotel} Best Time ({row["month"]})', edgecolor='black', s=100, zorder=5)

plt.title('Average Daily Rate (ADR) by Month, Year, and Hotel Type with Best Booking Times')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend(title='Hotel Type / Best Time', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
a['total_stay_nights'] = a['stays_in_weekend_nights'] + a['stays_in_week_nights']
 
# Remove rows where ADR is zero or less
data_cleaned = a[a['adr'] > 0]
 
# Further clean if needed: remove lengths of stay with very few records
stay_counts = data_cleaned['total_stay_nights'].value_counts()
valid_stay_lengths = stay_counts[stay_counts > 5].index
data_cleaned = data_cleaned[data_cleaned['total_stay_nights'].isin(valid_stay_lengths)]
 
# Analyze by hotel type
hotel_types = data_cleaned['hotel'].unique()
 
plt.figure(figsize=(14, 6))
 
for idx, hotel_type in enumerate(hotel_types):
    # Filter data for the specific hotel type
    hotel_data = data_cleaned[data_cleaned['hotel'] == hotel_type]
    # Group by total length of stay and calculate average ADR
    adr_by_stay_length = hotel_data.groupby('total_stay_nights')['adr'].mean().reset_index()
    # Find the optimal length of stay with the lowest ADR
    optimal_length_of_stay = adr_by_stay_length.loc[adr_by_stay_length['adr'].idxmin()]
    # Plot ADR vs. Length of Stay
    plt.subplot(1, len(hotel_types), idx + 1)
    sns.lineplot(data=adr_by_stay_length, x='total_stay_nights', y='adr')
    # Highlight the optimal length of stay
    plt.scatter(optimal_length_of_stay['total_stay_nights'], optimal_length_of_stay['adr'],
                color='red', s=100, edgecolor='black', zorder=5,
                label=f'Optimal Stay: {optimal_length_of_stay["total_stay_nights"]} nights')
    plt.title(f'Average Daily Rate (ADR) vs. Length of Stay ({hotel_type})')
    plt.xlabel('Total Length of Stay (Nights)')
    plt.ylabel('Average Daily Rate (ADR)')
    plt.legend()
    plt.grid(True)
 
plt.tight_layout()
plt.show()
```
