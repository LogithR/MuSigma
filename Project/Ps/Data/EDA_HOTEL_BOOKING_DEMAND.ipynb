```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure columns are of integer and string type as appropriate
a['arrival_date_year'] = a['arrival_date_year'].astype(int)
a['arrival_date_month'] = a['arrival_date_month'].astype(str)
a['arrival_date_day_of_month'] = a['arrival_date_day_of_month'].astype(int)

# Manually create a datetime column
a['arrival_date'] = pd.to_datetime(
    a[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].astype(str).agg('-'.join, axis=1),
    format='%Y-%B-%d'  # Adjust this format based on the month column's actual format
)

# Group by year and month to get the total number of bookings per month
demand_by_month = a.groupby(a['arrival_date'].dt.to_period('M')).size()

# Identify the peak month
peak_month = demand_by_month.idxmax()
peak_demand = demand_by_month.max()

# Plotting the demand by month with the peak month in red
plt.figure(figsize=(12, 6))

# Plot all months
colors = ['red' if month == peak_month else 'blue' for month in demand_by_month.index]
demand_by_month.plot(kind='bar', color=colors)

plt.title('Hotel Demand by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Ensure column names are correct based on your dataset
arrival_date_month = 'arrival_date_month'
reserved_room_type = 'reserved_room_type'
hotel = 'hotel'  # Replace with the actual column name if different

# Convert arrival_date_month to categorical type with proper ordering
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
a[arrival_date_month] = pd.Categorical(a[arrival_date_month], categories=months_order, ordered=True)

# Aggregate data: Count bookings by month and hotel type
monthly_bookings = a.groupby([arrival_date_month, reserved_room_type]).size().reset_index(name='booking_count')

# Ensure data is sorted correctly
monthly_bookings = monthly_bookings.sort_values([reserved_room_type, arrival_date_month])

# Create a pivot table for heatmap
pivot_table = monthly_bookings.pivot(index=arrival_date_month, columns=reserved_room_type, values='booking_count')

# Plotting the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Monthly Booking Counts by Room Type')
plt.xlabel('Reserved Room Type')
plt.ylabel('Month')
plt.show()
```

```python
# Average stay duration by room type
a['stay_duration'] = a['stays_in_weekend_nights'] + a['stays_in_week_nights']
avg_stay_duration = a.groupby('reserved_room_type')['stay_duration'].mean()
avg_stay_duration.plot(kind='bar', title='Average Stay Duration by Room Type')
plt.xlabel('Reserved Room Type')
plt.ylabel('Average Stay Duration (days)')
plt.show()
```
